```python
firstname = ["Attila" , "Dániel", "András", "Bálint", "Vivien", "László", "Kinga", "Adrienn", "Péter", "Sándor", "Árpád", "Anna", "Mária", "István", "Áron", "Barnabás", "Ferenc", "Lajos", "Márton", "Balázs", "Viktor", "Enikő", "Gábor", "Csaba"]
lastname = ["Nagy", "Kiss", "Szabó", "Kovács", "Horváth", "Balogh", "Kertész", "Györffy", "Tóth", "Héjas", "Part", "László", "Minden", "Barnabás", "Végh", "Áts", "Ocskó","Pocsai", "Dobák", "Fauszt", "Molnár","Kocsis", "Lakatos", "Pásztor", "Hegedűs", "Szűcs"]
import random as rnd
import csv
```

```python
matrix = []        #making the player list

for _ in range(4000):

    random_number = rnd.randint(1, 3)

    matrix.append([random_number , firstname[rnd.randint(0, len(firstname)-1)], lastname[rnd.randint(0, len(lastname)-1)] ])
    
for row in matrix:
    print(row)
```

```python
export = []

while len(matrix) > 4:
    helper = []
    count = 0
    index = 0
    while count < 1 and index < len(matrix):
        if matrix[index][0] == 1:
            helper=[matrix[index]]
            del matrix[index]
            count += 1
        else:
            index += 1
            if index == len(matrix):
                valt=rnd.randint(0, len(matrix)-1)
                helper=matrix[valt]
                del matrix[valt]
    count = 0
    index = 0
    while count < 1 and index < len(matrix):
        if matrix[index][0] == 2:
            helper[0].extend(matrix[index])
            del matrix[index]
            count += 1
        else:
            index += 1
            if index == len(matrix):
                valt=rnd.randint(0, len(matrix)-1)
                helper[0].extend(matrix[valt])
                del matrix[valt]
    count = 0
    index = 0
    while count < 1 and index < len(matrix):
        if matrix[index][0] == 3:
            helper[0].extend(matrix[index])
            del matrix[index]
            count += 1
        else:
            index += 1
            if index == len(matrix):
                valt=rnd.randint(0, len(matrix)-1)
                helper[0].extend(matrix[valt])
                del matrix[valt]
    for _ in range(2):
        valt=rnd.randint(0, len(matrix)-1)
        helper[0].extend(matrix[valt])
        del matrix[valt] 
    export.append(helper)
    
filename = "participants_export.csv"

with open(filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    
    for row in export:
        writer.writerow(row)

if len(matrix) > 0:
    print("Cannot be sorted into groups:")         #cannot be putted into a 5 people group
    print()
    for k in range(len(matrix)):
        print(matrix[k][1] +" "+ matrix[k][2])
```
